### Finding the Best Random State

- Before we proceed to model building we need to find the best random state for better accuracy
- we will use Linear Regression Algorithm to get the best random state
- We will use the metrics r2_score to get the accuracy of the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score

maxAccu=0
maxRS=0

for i in range(1, 1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=i)
    lr=LinearRegression()
    lr.fit(X_train, y_train)
    pred = lr.predict(X_test)
    r2 = r2_score(y_test, pred)
    if r2>maxAccu:
        maxAccu=r2
        maxRS=i

print("Best r2 score is", maxAccu,"on Random State", maxRS)
```

We will make a function to train test split in dependent and indepedent features and then train the model while predicting the 'AverageCost for two'
The function will caluclate the RMSE score , r2_score , calculates the Cross-Validation score and in the end it will Calculate the Difference between the r2_score and Cross-Validation score.

```python
def evaluate_regression_models(X, y, models):
   
    # Create DataFrame to store evaluation metrics
    metrics_df = pd.DataFrame(columns=["Model", "Best Random State", "MAE (Train)", "MAE (Test)", "MSE (Train)", "MSE (Test)", "R2 Score (Train)", "R2 Score (Test)", "CV Score"])
    
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=591)
    
    for model in models:
        # Initialize the model
        reg_model = model()
        
        # Fit the model
        reg_model.fit(X_train, y_train)
        
        # Make predictions on training and testing sets
        y_train_pred = reg_model.predict(X_train)
        y_test_pred = reg_model.predict(X_test)
        
        # Calculate evaluation metrics
        mae_train = mean_absolute_error(y_train, y_train_pred) * 100
        mae_test = mean_absolute_error(y_test, y_test_pred) * 100
        mse_train = mean_squared_error(y_train, y_train_pred) * 100
        mse_test = mean_squared_error(y_test, y_test_pred) * 100
        r2_train = r2_score(y_train, y_train_pred) * 100
        r2_test = r2_score(y_test, y_test_pred) * 100
        
        # Calculate cross-validation score
        cv_score = cross_val_score(reg_model, X, y, cv=5).mean()
        
        # Add results to metrics DataFrame
        metrics_df = metrics_df.append({
            "Model": reg_model.__class__.__name__,
            "Best Random State": 591,
            "MAE (Train)": mae_train,
            "MAE (Test)": mae_test,
            "MSE (Train)": mse_train,
            "MSE (Test)": mse_test,
            "R2 Score (Train)": r2_train,
            "R2 Score (Test)": r2_test,
            "CV Score": cv_score
        }, ignore_index=True)
        
    return metrics_df
```

```python
models = [LinearRegression, Ridge, Lasso, SVR, DecisionTreeRegressor, RandomForestRegressor, KNeighborsRegressor, AdaBoostRegressor, ExtraTreesRegressor]

# Evaluate models
results = evaluate_regression_models(X, y, models)
```

```python

def evaluate_classification_models(X, y, models):
    """
    X : Features DataFrame
    y : Target DataFrame
    models : List of classification models
    
    returns : DataFrame containing evaluation metrics for each model
    """
    # Create DataFrame to store evaluation metrics
    metrics_df = pd.DataFrame(columns=["Model", "Best Random State", "Accuracy (Train)", "Accuracy (Test)", "Precision (Train)", "Precision (Test)", "Recall (Train)", "Recall (Test)", "F1 Score (Train)", "F1 Score (Test)", "CV Score"])
    
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=121)
    
    for model in models:
        # Initialize the model
        clf_model = model()
        
        # Fit the model
        clf_model.fit(X_train, y_train)
        
        # Make predictions on training and testing sets
        y_train_pred = clf_model.predict(X_train)
        y_test_pred = clf_model.predict(X_test)
        
        # Calculate evaluation metrics
        acc_train = accuracy_score(y_train, y_train_pred)*100
        acc_test = accuracy_score(y_test, y_test_pred)*100
        precision_train = precision_score(y_train, y_train_pred, average='micro')
        precision_test = precision_score(y_test, y_test_pred, average='micro')
        recall_train = recall_score(y_train, y_train_pred, average='micro')
        recall_test = recall_score(y_test, y_test_pred, average='micro')
        f1_train = f1_score(y_train, y_train_pred, average='micro')
        f1_test = f1_score(y_test, y_test_pred, average='micro')
        
        # Calculate cross-validation score
        cv_score = cross_val_score(clf_model, X, y, cv=5).mean()*100
        
        # Add results to metrics DataFrame
        metrics_df = metrics_df.append({
            "Model": clf_model.__class__.__name__,
            "Best Random State": 121,
            "Accuracy (Train)": acc_train,
            "Accuracy (Test)": acc_test,
            "Precision (Train)": precision_train,
            "Precision (Test)": precision_test,
            "Recall (Train)": recall_train,
            "Recall (Test)": recall_test,
            "F1 Score (Train)": f1_train,
            "F1 Score (Test)": f1_test,
            "CV Score": cv_score
        }, ignore_index=True)
        
    return metrics_df
```

```python
models = [LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, KNeighborsClassifier, SVC]
results = evaluate_classification_models(X, y, models)
results
```
